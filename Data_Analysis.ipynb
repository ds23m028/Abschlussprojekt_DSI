```python
from pymongo import MongoClient
```

```python
try:
    client = MongoClient("mongodb://localhost:27017/")
    print("connected successfully")
except:
    print ("could not connect to mongoDB")
```

```python
db_Mongo = client.get_database('DSI_Abschlussprojekt')
```

```python
collection= db_Mongo.Geburten_nach_Einkommen_Wien
```

```python
list(collection.find())
```

```python
query = {
    "Total_Salary": {"$gt": 10000.0},
    "Births": {"$gt": 100.0}
}
```

```python
result = collection.find_one(query)
```

```python
if result:
    print("District_code:", result["District_code"])
else:
    print("No matching listing found.")
```

```python
collection.distinct("District_code")
```
